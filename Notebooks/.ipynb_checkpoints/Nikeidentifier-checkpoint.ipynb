## **2. Dataset Preparation & Preprocessing 🏗️**
### **Data Structure**
- Dataset is divided into **Nike** and **Other** categories.
- Each category has **multiple shoe models**.
- Images were filtered based on:
  - **Resolution (> 50x50 pixels)**
  - **Blur detection (Laplacian variance)**
  - **Corrupt image removal**

```python
# Function to validate images before processing
def is_image_valid(image_path, min_resolution=(50, 50), blur_threshold=100):
    try:
        # Load image using PIL
        with Image.open(image_path) as img:
            img.verify()  # Verify image integrity
        
        img = Image.open(image_path).convert("RGB")

        # Check resolution
        if img.width < min_resolution[0] or img.height < min_resolution[1]:
            print(f"Skipping: {image_path} - Resolution too low ({img.width}x{img.height}).")
            return False

        # Convert to grayscale for blurriness detection
        gray = img.convert("L")
        laplacian_var = cv2.Laplacian(np.array(gray), cv2.CV_64F).var()

        if laplacian_var < blur_threshold:
            print(f"Skipping: {image_path} - Too blurry (Variance: {laplacian_var:.2f}).")
            return False

        return True  # Image is valid
    except UnidentifiedImageError:
        print(f"Skipping: {image_path} - Unidentified image format.")
        return False
    except Exception as e:
        print(f"Skipping: {image_path} - Error: {e}")
        return False
```

```python
# Function to preprocess and save images with unique filenames
def preprocess_and_save(image_path, destination_path, resize_shape=(224, 224)):
    try:
        img = Image.open(image_path).convert("RGB")
        img = img.resize(resize_shape, Image.Resampling.LANCZOS)

        os.makedirs(destination_path, exist_ok=True)

        # Ensure unique filenames by appending a UUID
        base_filename = os.path.splitext(os.path.basename(image_path))[0]
        unique_filename = f"{base_filename}_{uuid.uuid4().hex}.jpg"
        save_path = os.path.join(destination_path, unique_filename)

        img.save(save_path, format="JPEG", quality=95)
        return save_path
    except Exception as e:
        print(f"Error processing {image_path}: {e}")
        return None
```

```python
# Process images in subfolders and handle duplicate names
def process_images_in_subfolders(root_folder, destination_folder):
    count = 0
    for subdir, _, files in os.walk(root_folder):
        for filename in files:
            image_path = os.path.join(subdir, filename)

            if filename.startswith(".") or not os.path.isfile(image_path):
                continue  

            if is_image_valid(image_path):
                saved_path = preprocess_and_save(image_path, destination_folder)
                if saved_path:
                    count += 1
                    print(f"Saved: {saved_path}")
    print(f"✅ Processed {count} images into {destination_folder} as JPEGs.")
```

```python
# Organize Nike and Other images
process_images_in_subfolders(nike_path, nike_organized_path)
process_images_in_subfolders(other_path, other_organized_path)

print(f"\nOrganized dataset is ready at: {organized_dataset_path}")
```

### **Preprocessing Steps**
- **Resizing** images to **224x224** for MobileNetV2.
- **Augmentation Techniques**:
  - Rotation (40°)
  - Brightness shifts
  - Zoom (up to 50%)
  - Horizontal flips
  - Channel shifts
  - Shear transformations

```python
# Define ImageDataGenerator for CPU optimization
data_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    brightness_range=[0.7, 1.3],
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.5,
    channel_shift_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)
# Create training and validation generators
train_generator = data_gen.flow_from_directory(
    organized_dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

val_generator = data_gen.flow_from_directory(
    organized_dataset_path,
    target_size=(224, 224),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
# Load pre-trained MobileNetV2 model
base_model = MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights="imagenet")
base_model.trainable = True
for layer in base_model.layers[:-40]:
    layer.trainable = False  
```

```python
# Define the classification model
model = models.Sequential([
    base_model,
    layers.BatchNormalization(),
    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu', kernel_regularizer=regularizers.l2(0.002)),
    layers.Dropout(0.4),
    layers.Dense(256, activation='relu', kernel_regularizer=regularizers.l2(0.002)),
    layers.Dropout(0.4),
    layers.Dense(1, activation='sigmoid')
])
```

```python
# Compile the model with a learning rate scheduler
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0003)
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python

# Define callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.4, patience=4, min_lr=1e-6),
    tf.keras.callbacks.ModelCheckpoint("/Users/mani/nike_best_model.keras", monitor='val_loss', save_best_only=True)
]
```

```python
import gc
tf.keras.backend.clear_session()
gc.collect()
```

```python
# Train the model
epochs = 20
history = model.fit(train_generator, validation_data=val_generator, epochs=epochs, callbacks=callbacks)

print("Model training complete.")
```

```python
# Save the trained model
model.save("/Users/mani/Sneakers/nike_shoe_classifier.keras")
print("Model saved successfully.")
```

```python
# Function to test a folder of images and count Nike vs Other classifications
def predict_folder(folder_path, model, threshold=0.5):
    nike_count = 0
    other_count = 0
    nike_images = []
    other_images = []
    
    for filename in os.listdir(folder_path):
        image_path = os.path.join(folder_path, filename)
        if not os.path.isfile(image_path):
            continue

        try:
            img = Image.open(image_path).convert("RGB").resize((224, 224))
            img_array = np.array(img) / 255.0  # Normalize
            img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
            prediction = model.predict(img_array)[0][0]
            
            if prediction <= threshold:
                nike_count += 1
                nike_images.append(filename)
            else:
                other_count += 1
                other_images.append(filename)
        except Exception as e:
            print(f"Error processing {filename}: {e}")
    
    print(f"Nike Count: {nike_count}")
    print(f"Other Count: {other_count}")
    print("Nike Images:", nike_images)
    print("Other Images:", other_images)
```
